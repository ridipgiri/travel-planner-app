```python
import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from peft import LoraConfig, get_peft_model, TaskType

# 1. Load the synthetic dataset from the previous step
print("Loading dataset...")
dataset = load_dataset("json", data_files="email_finetuning_data.jsonl", split="train")

# 2. Load the Tokenizer and Base Model
model_id = "google/flan-t5-small"
print(f"Loading tokenizer and model: {model_id}...")
tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForSeq2SeqLM.from_pretrained(model_id)

# 3. Define the tokenization function
def tokenize_function(examples):
    # T5 works best with a specific task prefix
    inputs = ["Rewrite this email to be polite and professional: " + text for text in examples["input"]]

    # Tokenize the input texts
    model_inputs = tokenizer(inputs, max_length=128, truncation=True, padding="max_length")

    # Tokenize the target outputs (the polite rewrites)
    labels = tokenizer(text_target=examples["output"], max_length=128, truncation=True, padding="max_length")

    # PyTorch loss functions ignore index -100. We replace pad tokens with -100
    # so the model doesn't get penalized for padding predictions.
    labels_with_ignore_index = []
    for label in labels["input_ids"]:
        labels_with_ignore_index.append([l if l != tokenizer.pad_token_id else -100 for l in label])

    model_inputs["labels"] = labels_with_ignore_index
    return model_inputs

# Apply tokenization to the dataset
print("Tokenizing data...")
tokenized_dataset = dataset.map(tokenize_function, batched=True, remove_columns=["input", "output"])

# 4. Configure PEFT (LoRA)
# Flan-T5 is a Seq2Seq model, so we must specify TaskType.SEQ_2_SEQ_LM
lora_config = LoraConfig(
    r=8,                     # Rank of the update matrices (smaller = fewer parameters)
    lora_alpha=16,           # Scaling factor
    target_modules=["q", "v"], # Applying LoRA to the query and value attention matrices
    lora_dropout=0.05,       # Dropout probability for LoRA layers
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)

# Wrap the base model with the PEFT configuration
peft_model = get_peft_model(model, lora_config)

print("\n--- PEFT Model Setup Complete ---")
peft_model.print_trainable_parameters()
```

```python
from transformers import TrainingArguments, Trainer, DataCollatorForSeq2Seq

# 5. Configure Training Arguments
training_args = TrainingArguments(
    output_dir="./results",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir="./logs",
    logging_steps=10,
    learning_rate=2e-4,
    save_strategy="epoch",
    report_to="none"
)

# Data collator for sequence-to-sequence models
data_collator = DataCollatorForSeq2Seq(tokenizer, model=peft_model)

# 6. Initialize Trainer
trainer = Trainer(
    model=peft_model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator
)

# 7. Train the model
print("\n--- Starting Training ---")
trainer.train()
print("\n--- Training Complete ---")
```

```python
# 1. Define a brand new rude email to test
test_email = "Why didn't you finish this yesterday? You are so slow and I am tired of waiting."

# 2. Format the input exactly as we did during training
prompt = "Rewrite this email to be polite and professional: " + test_email

# 3. Tokenize the input and move it to the same device as the model (GPU)
inputs = tokenizer(prompt, return_tensors="pt").to(peft_model.device)

# 4. Generate the output
print("Generating response...\n")
outputs = peft_model.generate(
    **inputs,
    max_new_tokens=50,       # Maximum length of the generated polite email
    temperature=0.7,         # Controls randomness (lower is more focused)
    do_sample=True
)

# 5. Decode and print the result
polite_rewrite = tokenizer.decode(outputs[0], skip_special_tokens=True)

print(f"**Original:** {test_email}")
print(f"**Polite Rewrite:** {polite_rewrite}")
```
